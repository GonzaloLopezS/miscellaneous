```python
def gen_punto()-> Tuple[float,float]:
  x: float = random.uniform(-1,1)
  y: float = random.uniform(-1,1)
  return x,y
```

```python
def funcion_mapeo(coord:Tuple[float,float])-> int:
  r = math.sqrt(coord[0]**2+coord[1]**2)
  return int(r < 1)
```

```python
def funcion_mapeo(row:str)-> Dict[str,int]:
  lista_palabras: Dict[str,int] = {}
  palabras = row.replace(","," ").split()
  for palabra in palabras:
    if palabra in lista_palabras:
      lista_palabras[palabra] += 1
    else:
      lista_palabras[palabra] = 1

  return lista_palabras
```

```python
entrada_1=funcion_mapeo("hola mundo")
entrada_1
```

```python
def funcion_reduccion(param_a:Dict[str,int],param_b:Dict[str,int])->Dict[str,int]:
  diccionario_palabras = param_b

  for palabra in param_a.keys():
    if palabra in param_b.keys():
      diccionario_palabras[palabra] = param_a[palabra] + param_b[palabra]
    else:
      diccionario_palabras[palabra] = param_a[palabra]

  return diccionario_palabras
```

```python
funcion_reduccion(entrada_2,entrada_1)
```

```python
reduced= reduce(funcion_reduccion,map_result)
```

```python
sorted_dict = {key:value for key, value in sorted(reduced.items(), key=lambda reduced: reduced[1], reverse=True)}

print(sorted_dict)
# reduced.sort()
```
